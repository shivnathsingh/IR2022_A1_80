```python
def Serach_word_inDict(word):
  list1=[]
  if (Posting_dict.get(word, 'Not Present') == 'Not Present'):
    list1.append([])
  else:
    list1.append(Posting_dict[word]) #Posting_dict also return a list so now  list1 return as a list of list
  return list1
```

```python
def And_fun( l1, l2):
  solution=[]
  i=0
  j=0
  comparison =0
  while(i<len(l1) and j<len(l2)):
    comparison+=1
    if(l1[i]==l2[j]):
      solution.append(l1[i])
      i=i+1
      j=j+1
          
    elif(l1[i]>l2[j]):
      j=j+1
    else:
      i=i+1   
  return solution,comparison;    





```

```python
def OR_fun(l1,l2):
  solution=[]
  i=0
  j=0
  comparision =0
  while(i<len(l1) and j<len(l2)):
    comparision+=1 
    if(l1[i]==l2[j]):
      solution.append(l1[i])
      i=i+1
      j=j+1
    elif(l1[i]<l2[j]):
      solution.append(l1[i])
      i=i+1
    else:
      solution.append(l2[j])
      j+=1    
  while(j<len(l2)):
    solution.append(l2[j]);
    j+=1
  while(i<len(l1)):
    solution.append(l1[i])
    i+=1;  
  solution.sort(key=int)
  return solution,comparision;    
```

```python
def AND_NOT_fun(l1,l2):
  comparision=0
  i=0
  j=0
  solution =[]
  while(i<len(l1) and j<len(l2)):
    comparision+=1
    if(l1[i]==l2[j]):
      i+=1
      j+=1
    elif(l1[i]<l2[j]):
      solution.append(l1[i])
      i+=1 
    else:
      j+=1
  while(i<len(l1)):
    solution.append(l1[i])
    i+=1
  return solution,comparision
```

```python
def not_fun(l2):
  list1=[]
  for i in range(1,1134):
    if i not in l2:
      list1.append(i)
  return list1    
```

```python
def OR_NOT_fun(l1,l2):
  not_in=[]
  not_in =not_fun(l2);
  solution,com=OR_fun(l1,not_in);
  return solution,com


```

```python

def Query_preprocessing(input,ope):
  list=[]
  text=input.lower()   #to convert in to lower case 
  text = re.sub(r'[^\w\s]', ' ', text)  # To remove puntuation
  word_tokens = word_tokenize(text)   # toknization 
  final_tokens=[w for w in word_tokens if w not in stop_words] # stopword remove
  final_tokens=[ lemmatizer.lemmatize(word) for word in final_tokens]
  ope=ope.upper()
  ope = re.sub(r'[^\w\s]',' ', ope)
  ope_token=word_tokenize(ope)
  
  i=0
  for w in final_tokens:
    list.append(w)
    if(i<len(ope_token)):
      list.append(ope_token[i]);
      i=i+1;
  return list,final_tokens,ope_token;

```
